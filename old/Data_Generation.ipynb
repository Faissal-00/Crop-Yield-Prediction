```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Define parameters
years = np.arange(2018, 2025)
crop_types = ['Wheat', 'Corn', 'Rice']
regions = ['North', 'South', 'West']

# Number of samples
num_samples = 20000  # Adjust as needed

# Generate synthetic data
data = {
    'Year': np.random.choice(years, num_samples),
    'Crop Type': np.random.choice(crop_types, num_samples),
    'Yield (kg/ha)': np.random.randint(4500, 8500, num_samples),
    'Avg Temp (°C)': np.round(np.random.uniform(17, 30, num_samples), 1),
    'Total Rainfall (mm)': np.random.randint(280, 410, num_samples),
    'Avg Sunlight (hours/day)': np.round(np.random.uniform(6, 12, num_samples), 1),
    'Soil pH': np.round(np.random.uniform(5.8, 6.8, num_samples), 1),
    'Soil Moisture (%)': np.round(np.random.uniform(24, 45, num_samples), 1),
    'Fertilizer Used (kg/ha)': np.random.randint(95, 160, num_samples),
    'Pesticide Used (L/ha)': np.round(np.random.uniform(4.5, 7.0, num_samples), 1),
    'Region': np.random.choice(regions, num_samples)
}

# Create DataFrame
df = pd.DataFrame(data)

# Display first 20 rows
df.head()
```

```python
df.to_csv('CropYieldData.csv')
```
